### ‚öñÔ∏è **2. Boxplot**

Un diagrama de caja y bigotes muestra la distribuci√≥n de datos cuantitativos de forma que facilita las comparaciones entre variables categ√≥rica. Muestra los cuartiles del conjunto de datos, mientras que los bigotes se extienden para mostrar el resto de la distribuci√≥n, excepto los puntos considerados "at√≠picos".

```python
# Peso por especie
plt.figure(figsize=(7, 4))

sns.boxplot(data=df_penguins,
            y='body_mass_g',
            hue='species',
            palette="Set2", )

plt.title('Distribuci√≥n de masa corporal por especie')
plt.xlabel('Especie')
plt.ylabel('Masa corporal (g)')
```

### ‚öñÔ∏è **3. Violinplot**

Un diagrama de viol√≠n cumple una funci√≥n similar a un diagrama de caja y bigotes. Muestra la distribuci√≥n de los puntos de datos tras agruparlos por una (o m√°s) variables. A diferencia de un diagrama de caja, cada viol√≠n se dibuja utilizando una estimaci√≥n de densidad.


```python
# Pico largo vs. pico profundo por especie
plt.figure(figsize=(8, 4))
sns.violinplot(data=df_penguins,
               y='flipper_length_mm',
               hue='species',
               palette="husl")
plt.title('Distribuci√≥n de la longitud de aleta por especie')
plt.xlabel('Especie')
plt.ylabel('Masa corporal (g)')
```

### üí°  **6. Relaci√≥n**

Crea un gr√°fico de dispersi√≥n con una l√≠nea de regresi√≥n superpuesta, visualizando la relaci√≥n lineal entre dos variables. Permite mostrar m√∫ltiples gr√°ficos de regresi√≥n en diferentes subconjuntos de datos. Esta funci√≥n es especialmente √∫til para explorar c√≥mo var√≠an las relaciones entre categor√≠as definidas por otras variables.


```python
#Relaci√≥n entre largo de aleta y largo del pico, por especie
plt.figure(figsize=(8, 5))
sns.lmplot(
    data=df_penguins,
    x='flipper_length_mm',
    y='bill_length_mm',
    hue="species" # hue sirve para agrupar en funcion a una variable.
)

plt.title('Relaci√≥n lineal entre longitud de aleta y pico')
plt.xlabel('Longitud de aleta (mm)')
plt.ylabel('Longitud del pico (mm)')
plt.show()
```

```python
#Relaci√≥n entre largo de aleta y peso corporal, por especie, isla y sexo
sns.lmplot(
    data=df_penguins,
    x='flipper_length_mm',
    y='bill_length_mm',
    hue="species", # hue sirve para agrupar en funcion a una variable.
    col="island", # col separa en x columnas los graficos en funci√≥n a una variable categorica.
    row="sex", # row separa en x filas los graficos en funci√≥n a una variable categorica.
    palette="Set2"
)
```

### üí°  **7. Heatmaps**

Es otra herramienta poderosa de Seaborn que permite visualizar simult√°neamente:


*   Un diagrama de dispersi√≥n (scatterplot) entre dos variables.

*   Junto con sus distribuciones marginales (histogramas o KDE) en los ejes X e Y.


```python
#Crear un heatmap de correlaci√≥n entre variables cuantitativas del dataset penguins para:
#Ver qu√© medidas est√°n m√°s relacionadas entre s√≠.
#Ilustrar c√≥mo interpretar correlaciones entre variables biol√≥gicas (como si fueran biomarcadores o genes).

# Eliminar filas con valores faltantes
df_penguins = df_penguins.dropna()

# Seleccionar solo columnas num√©ricas
numeric_data = df_penguins.select_dtypes(include='number')

# Calcular la matriz de correlaci√≥n
correlation_matrix = numeric_data.corr()

# Crear el heatmap
plt.figure(figsize=(8, 4))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Matriz de correlaci√≥n entre variables morfol√≥gicas de ping√ºinos')
plt.show()
```

```python
#heatmap de conteo por isla y especie
# Eliminar valores faltantes
# Eliminar filas con NaN en columnas clave
df_clean = df_penguins.dropna(subset=['species', 'island'])

# Crear tabla de conteo (frecuencia absoluta)
tabla = pd.crosstab(df_clean['island'], df_clean['species'])

# Dibujar heatmap
sns.heatmap(tabla, annot=True, cmap='YlOrBr')
plt.title('Abundancia por Isla y Especie')
plt.xlabel('Especie')
plt.ylabel('Isla')
plt.tight_layout()
plt.show()
```

#### **5. ¬øCu√°les son los 25 generos de bacterias m√°s abundantes y su proporci√≥n por grupo de paciente?**
**Prompt:** Generar una gr√°fica de barras apiladas de los distintos grupos de pacientes de pacientes, de tal manera que el eje Y muestre la abundancia relativa de TODOS g√©neros bacterianos por grupo. Ees decir, las abundancias deben ser normalizadas por grupo para que sumen 100%.
Considerar:
* La columna 1 tiene el grupo que indica el tipo de paciente.
* Las columnas de la 2 a la 77 (√≠ndices 1 al 76) contienen las abundancias de distintos g√©neros bacterianos.
Usar seaborn.


```python
# prompt: Generar una gr√°fica de barras apiladas de los distintos grupos de pacientes de pacientes, de tal manera que el eje Y muestre la abundancia relativa de TODOS g√©neros bacterianos por grupo. Ees decir, las abundancias deben ser normalizadas por grupo para que sumen 100%.
# Considerar:
# * La columna 1 tiene el grupo que indica el tipo de paciente.
# * Las columnas de la 2 a la 77 (√≠ndices 1 al 76) contienen las abundancias de distintos g√©neros bacterianos.
# Usar seaborn.

import matplotlib.pyplot as plt
# Seleccionar solo las columnas de abundancia y el grupo
df_abundance = df_microbiome.iloc[:, [0] + list(range(2, 77))].copy()

# Calcular la abundancia total por grupo
df_abundance['total_abundance'] = df_abundance.iloc[:, 1:].sum(axis=1)

# Normalizar las abundancias por grupo
for col in df_abundance.columns[1:-1]:
    df_abundance[col] = (df_abundance[col] / df_abundance['total_abundance']) * 100

# Eliminar la columna de abundancia total
df_abundance = df_abundance.drop(columns=['total_abundance'])

# Calcular la abundancia promedio de cada g√©nero por grupo
df_grouped_abundance = df_abundance.groupby('grupo').mean()

# Transponer el DataFrame para que los g√©neros sean filas y los grupos columnas
df_grouped_abundance_T = df_grouped_abundance.T

# Ordenar los g√©neros por abundancia total descendente para seleccionar los m√°s abundantes
top_genres = df_grouped_abundance_T.sum(axis=1).sort_values(ascending=False).index

# Seleccionar un n√∫mero razonable de los g√©neros m√°s abundantes para la visualizaci√≥n
# Puedes ajustar este n√∫mero (por ejemplo, 15 o 20)
num_top_genres = 25
df_top_genres = df_grouped_abundance_T.loc[top_genres[:num_top_genres]]

# Preparar los datos para el gr√°fico de barras apiladas
df_top_genres_melted = df_top_genres.reset_index().melt(id_vars='index', var_name='grupo', value_name='relative_abundance')

# Renombrar la columna de √≠ndice a 'genero'
df_top_genres_melted = df_top_genres_melted.rename(columns={'index': 'genero'})


# Crear la gr√°fica de barras apiladas
plt.figure(figsize=(12, 8))
sns.barplot(data=df_top_genres_melted,
            x='grupo',
            y='relative_abundance',
            hue='genero',
            dodge=False) # Dodge=False para que sean barras apiladas

plt.title('Abundancia Relativa de los Principales G√©neros Bacterianos por Grupo de Paciente')
plt.xlabel('Grupo de Paciente')
plt.ylabel('Abundancia Relativa (%)')
plt.xticks(rotation=0) # Asegura que las etiquetas del eje X no giren si son cortas
plt.legend(title='G√©nero Bacteriano', bbox_to_anchor=(1.05, 1), loc='upper left') # Coloca la leyenda fuera del gr√°fico
plt.tight_layout() # Ajusta el dise√±o para evitar que los elementos se solapen
plt.show()
```

```python
# prompt: Generar una gr√°fica de boxplot de la diversidad calculada por el indice de shannon por grupo de paciente.
# Considerar:
# * La columna 1 tiene el grupo que indica el tipo de paciente.
# * Las columnas de la 2 a la 77 (√≠ndices 1 al 76) contienen las abundancias de distintos g√©neros bacterianos.
# Usar seaborn.

import matplotlib.pyplot as plt
import numpy as np
# Calcular la diversidad de Shannon para cada paciente
# Definir una funci√≥n para calcular el √≠ndice de Shannon
def shannon_diversity(abundances):
  # Eliminar las abundancias de cero antes del c√°lculo para evitar log(0)
  abundances = abundances[abundances > 0]
  total_individuals = abundances.sum()
  proportions = abundances / total_individuals
  shannon_index = -np.sum(proportions * np.log(proportions))
  return shannon_index

# Aplicar la funci√≥n de Shannon a las columnas de abundancia para cada fila (paciente)
df_microbiome['shannon_diversity'] = df_microbiome.iloc[:, 2:77].apply(shannon_diversity, axis=1)

# Crear la gr√°fica de boxplot de la diversidad de Shannon por grupo de paciente
plt.figure(figsize=(8, 4))
sns.boxplot(data=df_microbiome, x='grupo', y='shannon_diversity', palette='viridis')
plt.title('Diversidad de Shannon por Grupo de Paciente')
plt.xlabel('Grupo de Paciente')
plt.ylabel('√çndice de Shannon')
plt.show()
```
